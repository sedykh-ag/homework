# Интерполяция полиномом Лагранжа.

Напишите класс, который конструирует интерполяционный полином Лагранжа, проходящий через точки, заданные как `xk` и `yk`.

```python
import numpy as np

class LagrangeInterpolator:
    """Lagrange interpolating polynomial.
    
    Given a set of pairs ``(x_k, y_k)``, construct 
    a Lagrange polynomial ``f(x)``, such that
    
    .. math::

        f(x_k) = y_k   for k =0, ..., n-1
    
    Parameters
    ----------
    xk : array_like, shape(n,)
        Abscissas
    yk : array_like, shape(n,)
        Ordinates
    
    Attributes
    ----------
    __call__
    
    """
    def __init__(self, xk, yk):
        self.xk = np.asarray(xk, dtype=float)
        self.yk = np.asarray(yk, dtype=float)
        
    def __call__(self, x):
        """Evaluate the interpolator at a given point.
        
        Parameters
        ----------
        x : float
        
        Returns
        -------
        the value of the interpolator at ``x``.
        """
        L = 0
        n = self.xk.shape[0]
        
        for i in range(n):
            x_i = self.xk[i]
            y_i = self.yk[i]
            
            xk_i = np.hstack([ self.xk[:i], self.xk[i+1:] ]) # xk array without x_i element
            l_i = np.prod( (x - xk_i) / (x_i - xk_i) )
            L += y_i * l_i
        
        return L
```

```python
import matplotlib.pyplot as plt
from scipy import interpolate

x1 = np.array([-2, -1,  0,  1,  2])
y1 = np.array([ 1,  0,  0,  0, -1])

x0 = np.linspace(-3, 3, 100)

LI = LagrangeInterpolator(x1, y1)
LL = interpolate.lagrange(x1, y1)

plt.scatter(x1, y1, color='k')
plt.plot(x0, [LI(x) for x in x0], color='k', label="LagrangeInterpolator")
plt.plot(x0, LL(x0), linewidth=10, alpha=0.25, color='g', label="scipy.interpolate.lagrange")
plt.xlim(-3, 3)
plt.ylim(-2, 2)
plt.legend()
plt.show()
```

```python
def runge_func(x, a=25):
    return 1.0 / (1.0 + a*x**2)

xx = np.linspace(-2, 2, 21)
yy = runge_func(xx)

lagr = LagrangeInterpolator(xx, yy)

from numpy.testing import assert_allclose

assert_allclose(yy,
                [lagr(xval) for xval in xx],
                atol=1e-14)
```

## Феномер Рунге

Рассмотрим функцию Рунге, $y = 1/(1 + 25x^2)$. Интерполируйте эту функцию на интервале $x\in [-2, 2]$, используя полином Лагранжа с $m=3, 5, 7, 11$. Используйте равномерную сетку. Нарисуйте результат интерполяции вместе с исходной функцией на одном графике. 

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

n = [4, 6, 8, 12] # m + 1 points
fig, axs = plt.subplots(2, 2, figsize=(16, 10))

x0 = np.linspace(-2.5, 2.5, 100)

for i in range(4):
    x = np.linspace(-2, 2, n[i])
    y = 1 / (1 + 25 * x**2)
    L = interpolate.lagrange(x, y)
    
    ax = axs.flatten()[i]
    ax.scatter(x, y, color='k', label="data")
    ax.plot(x0, 1 / (1 + 25 * x0**2), color='g', label="true Runge")
    ax.plot(x0, L(x0), '--', color='k', label=f"L{n[i] - 1}")
    ax.legend()
    ax.set_ylim(-0.5, 1.5)

plt.show()
```

Допишите функцию, которая возвращает узлы Чебышева. 

```python
def cheb_nodes(n, a=-1, b=1):
    """
    Chebyshev nodes of degree $n$ on $[a, b]$
    """
    return np.array([0.5*(a+b) + 0.5*(b-a)*np.cos((2*k-1)*np.pi / (2*n)) for k in range (n, 0, -1)])
```

```python
nodes_11 = cheb_nodes(11)
nodes_11 = np.asarray(nodes_11)
assert (nodes_11[1:] > nodes_11[:-1]).all()

from scipy.special import roots_chebyt
nodes, weights = roots_chebyt(5) 

assert_allclose(cheb_nodes(5),
                nodes, atol=1e-14)

assert_allclose(cheb_nodes(5, a=-1, b=3),
                nodes*2 + 1, atol=1e-14)
```

Другим известным трудным тестом для интерполяции является следующая периодическая функция:

$u(x)=\frac{\sqrt{\varepsilon(2+\varepsilon)}}{2 \pi(1+\varepsilon-\cos x)}, \quad-\pi \leq x \leq \pi, \quad \varepsilon=0.21  $

Проведите аналогичные вычисления и постройте графики для неё.

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
from scipy.special import roots_chebyt

eps = 0.21
f = lambda x: np.sqrt(eps*(2+eps)) / (2*np.pi*(1+eps-np.cos(x))) # u(x)

n = [4, 6, 8, 12] # m + 1 points
fig, axs = plt.subplots(2, 2, figsize=(16, 10), dpi=250)

x0 = np.linspace(-np.pi - 0.5, np.pi + 0.5, 100)    # сетка для визуализации

for i in range(4):
    x_c = cheb_nodes(n[i], -np.pi, np.pi)   # чебышевская сетка (c = chebyshev)
    x_u = np.linspace(-np.pi, np.pi, n[i])  # равномерная сетка (u = uniform)
    y_c = f(x_c)
    y_u = f(x_u)
    L = interpolate.lagrange(x_c, y_c)    # Лагранж на чебышевской сетке
    L_0 = interpolate.lagrange(x_u, y_u)  # Лагранж на равномерной сетке
    S = interpolate.CubicSpline(x_u, y_u) # Кубический сплайн на равномерной сетке
    
    ax = axs.flatten()[i]
    ax.scatter(x_u, y_u, c='k', marker='o', label="uniform")
    ax.scatter(x_c, y_c, c='k', marker='x', label="chebyshev")
    ax.plot(x0, f(x0), color='g', label="true")
    ax.plot(x0, L_0(x0), '--', color='r', label=f"L{n[i] - 1} uni")
    ax.plot(x0, L(x0), '--', color='k', label=f"L{n[i] - 1} cheb")
    ax.plot(x0, S(x0), linewidth=5, alpha=0.3, color='b', label="Spline uni")
    
    ax.legend()
    ax.set_ylim(-0.25, 1.0)

plt.show()
```

Построим зависимость ошибки от количества узлов

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

def estimate_error(f, n):
    x0 = np.linspace(-1, 1, 400)
    CH = np.polynomial.chebyshev.Chebyshev.interpolate(func=f, deg=n)
    return np.max(np.abs(f(x0) - CH(x0)))

f1 = lambda x: np.sin(6*x) + np.sin(60*np.e**x)
f2 = lambda x: 1 / (1 + 1000*(x+0.5)**2) + 1 / np.sqrt(1 + 1000*(x-0.5)**2)

nodes = np.arange(1, 201)

fig, axs = plt.subplots(1, 2, figsize=(16, 5), dpi=150)
for i in range(2):
    f = [f1, f2][i]
    err = [estimate_error(f, n) for n in nodes]
    
    ax = axs[i]
    ax.plot(nodes, err, color='r', label=f"f{i+1} err")
    ax.set_xlabel("nodes")
    ax.set_ylabel("err")
    ax.legend()

plt.show()
```

$$
y_{1}(x)=p_{0}+p_{1} x+p_{2} x^{2}+p_{3} x^{3}+p_{4} x^{4}+p_{5} x^{5}, y_{2}(x)=\frac{p_{0}+p_{1} x+p_{2} x^{2}+p_{3} x^{3}}{1+q_{1} x+q_{2} x^{2}}
$$


Какая из этих формул лучше описывает $y(x)$ на заданном интервале? Постройте график экстраполяции $y(x)$ за пределы исходных данных. Можете ли вы угадать аналитическое выражение для $y(x) ?$

```python
import numpy as np

data = np.load("data_7_1.npz")
x4, y4 = data["x4"], data["y4"]
plt.scatter(x4, y4, s=5)
plt.show()
```

```python
from scipy.optimize import curve_fit

def f1(x, p0, p1, p2, p3, p4, p5):
    return p0 + p1 * x + p2 * x**2 + p3 * x**3 + p4 * x**4 + p5 * x**5

def f2(x, p0, p1, p2, p3, q1, q2):
    return (p0 + p1 * x + p2 * x**2 + p3 * x**3) / (1 + q1 * x + q2 * x**2)

popt1, *_ = curve_fit(f1, x4, y4)
popt2, *_ = curve_fit(f2, x4, y4)

x0 = np.linspace(-1.0, 1.0, 200)
plt.figure(figsize=(10, 8))
plt.scatter(x4, y4, s=10, color="k")
plt.plot(x0, f1(x0, *popt1), color="r", label="f1")
plt.plot(x0, f2(x0, *popt2), color="b", label="f2")
plt.legend()
plt.show()
```

```python
plt.plot(x4, f1(x4, *popt1)-y4, color="r", label="f1") 
plt.plot(x4, f2(x4, *popt2)-y4, color="b", label="f2") 
plt.legend()
plt.xlabel("x")
plt.ylabel("err")
plt.show()
```

По графику видно, что $f_2$ приближает лучше.
