# Task D3

$u_{xx}+u_{yy}=f(x,y)$ на многосвязной области интегрирования с нулевым начальным приближением и различными граничными условиями первого или второго типа.

Граничное условие можно придумать самостоятельно, как и форму контура. Метод решения любой, можно метод средних арифметических.

```python
import matplotlib.pyplot as plt
import numpy as np

N = 500
M = 500

h = 1 / max(N, M)

grid = np.zeros((N, M))

x1 = np.zeros(M) + 0
x2 = np.zeros(M) + 0
y1 = np.zeros(N) + 0
y2 = np.zeros(N) + 0

grid[0] = x1
grid[N-1] = x2
grid[:, 0] = y1
grid[:, M-1] = y2
```

```python
def f(x, y):
    A = (x - 0.5) ** 2 + (y - 0.5) ** 2 < 0.02**2
    return 50 *  np.array(A)
```

```python
eps = 1e-6
r = 1e10
X, Y = np.meshgrid(np.linspace(0, 1, M, endpoint=False), np.linspace(0, 1, N, endpoint=False))
plt.imshow((f(X, Y)))
```

```python
while r > eps:
    r = np.max(np.abs(grid[1:N-1, 1:M-1] - (grid[2:N, 1:M-1] + grid[0:N-2, 1:M-1] + grid[1:N-1, 2:M] + grid[1:N-1, 0:M-2])/4 - f(X[1:N-1, 1:M-1], Y[1:N-1, 1:M-1])*h**2 / 4 ))
    grid[1:N-1, 1:M-1] = (grid[2:N, 1:M-1] + grid[0:N-2, 1:M-1] + grid[1:N-1, 2:M] + grid[1:N-1, 0:M-2])/4 + f(X[1:N-1, 1:M-1], Y[1:N-1, 1:M-1])*h**2 / 4
```

```python
plt.imshow(grid[200:300, 200:300])
```

# Task D4

Требуется решить двумерное уравнение уравнение параболического типа: $u_t=u_{xx}+u_{yy}+f(x,y)$ на многосвязной области интегрирования с нулевым начальным условием $u(t=0,x,y)=u_0(x,y)=0$

```python
import matplotlib.pyplot as plt
import numpy as np

N = 500
M = 500
T_steps = 5000
T = 0.005
h = 1 / max(N, M)
dt = T / T_steps
u = np.zeros((T_steps, N, M))
ux1 = np.zeros(M) + 0
ux2 = np.zeros(M) + 0
uy1 = np.zeros(N) + 0
uy2 = np.zeros(N) + 0
u[0, 0] = ux1
u[0, N-1] = ux2
u[0, :, 0] = uy1
u[0, :, M-1] = uy2
```

```python
def f(x, y):
    A = (x - 0.5) ** 2 + (y - 0.5) ** 2 < 0.02**2
    return 50 *  np.array(A)
```

```python
X, Y = np.meshgrid(np.linspace(0, 1, M, endpoint=False), np.linspace(0, 1, N, endpoint=False))
plt.imshow((f(X, Y)),)
```

```python
for i in range(1, T_steps):
    u[i, 1:N-1, 1:M-1] = u[i-1, 1:N-1, 1:M-1] +dt * (f(X[1:N-1, 1:M-1], Y[1:N-1, 1:M-1]) + (u[i-1, 2:N, 1:M-1] + u[i-1, 0:N-2, 1:M-1]+u[i-1, 1:N-1, 2:M] + u[i-1, 1:N-1, 0:M-2] - 4 * u[i-1, 1:N-1, 1:M-1]) / h**2)
```
